```python
class CoffeeType:
    def __init__(self, name, water_needed, milk_needed, coffee_needed, price):
        self.name = name
        self.water_needed = water_needed
        self.milk_needed = milk_needed
        self.coffee_needed = coffee_needed
        self.price = price


class CoffeeMachine:
    def __init__(self):
        self.water = 1000  # in ml
        self.milk = 500   # in ml
        self.coffee = 200 # in grams
        self.money = 0

        # Define available coffee types
        self.menu = {
            "Espresso": CoffeeType("Espresso", 50, 0, 18, 2.5),
            "Latte": CoffeeType("Latte", 200, 150, 24, 3.5),
            "Cappuccino": CoffeeType("Cappuccino", 250, 100, 24, 4.0),
        }

    def display_menu(self):
        print("Menu:")
        for coffee_name, coffee in self.menu.items():
            print(f"{coffee_name}: ${coffee.price}")
    
    def check_resources(self, coffee_type):
        coffee = self.menu[coffee_type]
        if self.water < coffee.water_needed:
            print("Sorry, not enough water!")
            return False
        if self.milk < coffee.milk_needed:
            print("Sorry, not enough milk!")
            return False
        if self.coffee < coffee.coffee_needed:
            print("Sorry, not enough coffee!")
            return False
        return True
    
    def make_coffee(self, coffee_type):
        if self.check_resources(coffee_type):
            coffee = self.menu[coffee_type]
            self.water -= coffee.water_needed
            self.milk -= coffee.milk_needed
            self.coffee -= coffee.coffee_needed
            self.money += coffee.price
            print(f"Here is your {coffee_type}. Enjoy!")
        else:
            print("Cannot make coffee. Please refill the machine.")
    
    def refill(self):
        self.water = 1000
        self.milk = 500
        self.coffee = 200
        print("Machine refilled!")
    
    def show_resources(self):
        print(f"Water: {self.water}ml")
        print(f"Milk: {self.milk}ml")
        print(f"Coffee: {self.coffee}g")
        print(f"Money: ${self.money}")


# Main Program
machine = CoffeeMachine()

while True:
    print("\n1. Show Menu\n2. Make Coffee\n3. Refill Machine\n4. Show Resources\n5. Exit")
    choice = input("Select an option: ")
    if choice == "1":
        machine.display_menu()
    elif choice == "2":
        coffee_type = input("Enter coffee type (Espresso, Latte, Cappuccino): ")
        if coffee_type in machine.menu:
            machine.make_coffee(coffee_type)
        else:
            print("Invalid coffee type.")
    elif choice == "3":
        machine.refill()
    elif choice == "4":
        machine.show_resources()
    elif choice == "5":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```
